```python
%cd /
from glob import glob

# 이미지 경로 list로 넣기
train_img_list = glob('/content/dataset/train/images/*.jpg')
valid_img_list = glob('/content/dataset/valid/images/*.jpg')

# txt 파일에 write
with open('/content/dataset/train.txt', 'w') as f:
	f.write('\n'.join(train_img_list) + '\n')

with open('/content/dataset/val.txt', 'w') as f:
	f.write('\n'.join(valid_img_list) + '\n')
```

디렉토리를 루트로 바꾸고 glob 이라는 패키지를 import 한다음에 모든 이미지를 긁어올거임

/content/dataset/train/images/*.jpg 이 경로에 있는 모든이미지 개수를 확인

```python
!pip install PyYAML
```

```python
import yaml

with open('/content/dataset/data.yaml', 'r') as f:
  data = yaml.load(f, Loader=yaml.FullLoader)
  #json 부르는 형식과 똑같음

print(data)

data['train'] = '/content/dataset/train.txt'
data['val'] = '/content/dataset/val.txt'

with open('/content/dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)

print(data)
```

위에걸 실행시키면 원래 train/images 였는데 -> /content/dataset/train.txt 로 바꼈고,
valid/images 였는데 -> /content/dataset/val.txt로 바뀜
